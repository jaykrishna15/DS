```python
a=['madam',343,453]
a_1=[]
for i in a:
    i=str(i)
    i_1=i[::-1]
    
    if(i==i_1):
        a_1.append((i,'is a palidrome'))
        
    else:
        a_1.append((i,'Not a palindrome'))

print(a_1)
```

```python
a=['madam',343,453]          #input
a_1={}                       # create a new dict
for i in a:                  # iterating the elements
    i=str(i)                 # type conversion
    i_1=i[::-1]              # reverse
    
    if(i==i_1):              ## checking the elements are same 
        a_1.update({i:'is a palidrome'})        # then update in the new dict
        
    else:                   # not same
        a_1.update({i:'Not a palindrome'})      # then update in the the new dict

print(a_1)                 # printing the over input
```

### For vs While
1:- for is automaic, while requied manual initialization,stopping,condition and increment/decrement.  
2:- while loop is more prone to enter in infinite loop unlike for  
3:- sa all the llops that can be run in for loop can also  run in  while loop but the reverse is not true  
4:- for loop requies a proper end state whereas when the stop state isn't properly known, then we can use while loop

```python
num = 153
rev = 0
len_num = len(str(num))

# keep a copy of original num
temp = num

while temp != 0:
    rem = temp % 10
    rev += rem ** len_num
    temp = temp // 10

if num == rev:
    print(num, 'is an Armstrong number')
else:
    print(num, 'is not an Armstrong number')
```
